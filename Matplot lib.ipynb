# Pie Chart

```python
my_dict={'ravi%':[20,35,15,17,13]}
```

```python
a=pd.DataFrame(data=my_dict,index=['BTC','ETH','LTC',"XRP",'ADA'])
a
```

```python
a.plot.pie(y="ravi%", figsize=(5,5),autopct='%.2f %%',pctdistance=0.5)
plt.title('CRYPTO PORTFOLIO PIE CHART')
```

```python
explode=(0.1,0,0.1,0.1,0.1)
a.plot.pie(y="ravi%", figsize=(5,5),autopct='%.2f %%',pctdistance=0.5,explode=explode)
plt.title('CRYPTO PORTFOLIO PIE CHART')
```

# Histograms

```python
r=pd.read_csv('Downloads\crypto_daily_prices.csv')
```

```python
r
```

```python
x=r['BTC-USD Price'].mean()
a=r['BTC-USD Price'].sum()

#num_bins=40
plt.figure(figsize=(20,15))

r['BTC-USD Price'].plot.hist(bins= 100,alpha= 1,figsize= (12,6))
#plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: x= '+str(x)+ ',a= ' +str(a));
```

```python
x=r['BTC-USD Price'].std()
a=r['BTC-USD Price'].sum()

#num_bins=40
plt.figure(figsize=(20,15))

r['BTC-USD Price'].plot.hist(bins= 40,alpha= 1,figsize= (12,6),color='black')
#plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: x= '+str(x)+ ',a= ' +str(a));
```
