```python
htmlContent=r.text
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

```python
s=soup.find('tbody',class_='lister-list').find_all('tr')
```

```python
s
```

```python
r=soup.find('td',class_='titleColumn').find('a')
```

```python
s=soup.find('tbody',class_='lister-list').find_all('tr')
c=0
rating=[]
movies=[]
for i in s:
    c=c+1
    rating.append(i.find('td',class_="ratingColumn imdbRating").find("strong").text)
    movies.append(i.find('td',class_ = 'titleColumn').find('a').text)
    print(movies,rating)
```

```python
print(c)
```

```python
import pandas as pd
```

```python
x=pd.DataFrame(rating,movies)
```
