```python
x=int(input("Enter the number: "))
l=[2000,500,200,200,100,20,10,1]
a={}
for i in l:          #It contains all the value of l
    if x>=i:         # if x>=i e.g if 2500>=i which contains no. of elements
        a[i]=x//i    # a[i]  i contains indexing  3000//2000=1 
        x=x%i        # x=1000%i=1000%500=2
print(a)             # print({2000:1,500:2})
```
