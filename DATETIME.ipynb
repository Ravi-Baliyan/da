# 2. Datetime in action! Practical Example Part#1

```python
avocardo=pd.read_csv("Downloads\\avocado (1).csv")
```

```python
avocardo
```

```python
avocardo['Date']=pd.to_datetime(avocardo['Date'])
avocardo
```

```python
avocardo.set_index(keys=['Date'],inplace=True)
avocardo.head(5)
```

```python
avocardo.loc["2018-01-21"]
```

```python
avocardo.iloc[5]
```

```python
avocardo.sort_index(inplace=True)
avocardo.truncate("2017-01-01","2018-02-01")
```

```python
avocardo.index=avocardo.index + pd.DateOffset(months=12,days=30)
```

```python
avocardo
```

```python
avocardo.index=avocardo.index - pd.DateOffset(months=12,days=30)
avocardo
```

```python
avocardo.resample(rule='M').mean()
```

```python
avocardo.resample(rule='A').mean()
```

```python
avocardo.resample(rule='M').sum()
```

```python
avocardo.resample(rule='Q').max()
```

```python
low_price=avocardo["AveragePrice"].where(avocardo["AveragePrice"]<1.2)
low_price.resample('M')
```

```python
low_price=avocardo["AveragePrice"].where(avocardo["AveragePrice"]<1.3).resample('M').count()
low_price
```

```python
avocardo.reset_index(inplace=True)
```

```python
avocardo['Day']=avocardo['Date'].dt.day
avocardo
```

```python
avocardo['Month']=avocardo['Date'].dt.month
avocardo
```

```python
avocardo['Year']=avocardo['Date'].dt.year
avocardo
```

```python
avocardo.set_index(keys=["Date"],inplace = True)
avocardo
```

```python
avocardo.resample(rule='M').mean()
```

```python
avocardo.resample(rule='M').mean()['AveragePrice'].plot(figsize=(10,5),marker='*',color='black')
```

```python
avocardo.resample(rule='Q').mean()['AveragePrice'].plot(figsize=(10,5),marker='*',color='black')
```

```python
avocardo.resample(rule='A').mean()['AveragePrice'].plot(figsize=(10,5),marker='*',color='black')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
sns.violinplot(y="AveragePrice",x="type",data=avocardo,color='black')
```
