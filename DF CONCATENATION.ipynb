# DataFrame Concatenation

```python
import pandas as pd
```

```python
d={'Bank Client ID':['1','2','3','4','5'],
  'First Name':['Nancy','Alex','Shep','Max','Allen'],
  'Last Name':['Rob','Ali','George','Mitch','Steve']}
d1=pd.DataFrame(d,columns=['Bank Client ID','First Name','Last Name'])
d1
```

```python
d={'Bank Client ID':['6','7','9','10','15'],
  'First Name':['Bill','Dina','Sarah','Heather','Holly'],
  'Last Name':['Christian','Mo','Steve','Bob','Michelle']}
d2=pd.DataFrame(d,columns=['Bank Client ID','First Name','Last Name'])
d2
```

```python
x=pd.concat([d1,d2])
```

```python
x
```
