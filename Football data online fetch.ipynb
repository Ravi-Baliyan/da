```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://towardsdatascience.com/tagged/football-analytics'
r=requests.get(url)
```

```python
a=r.text
```

```python
a
```

```python
htmlContent=a
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

```python
x=soup.find('div',class_='iz ja jb l').text
```

```python
print(x)
```

```python
x=soup.find('div',class_='iz ja jb l').find_all('div')
```

```python
for i in x:
    print(i.prettify)
```
